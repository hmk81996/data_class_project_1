```python
import ast

# Sample DataFrame creation for demonstration
# df = pd.DataFrame({'amenities': [...]})  # Your DataFrame

# Step 1: Convert the string representation of the list to an actual list safely
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return []  # Return an empty list if there's an error

df['amenities'] = df['amenities'].apply(safe_literal_eval)

# Step 2: Get unique amenities
unique_amenities = list(set(amenity for sublist in df['amenities'] for amenity in sublist))

# Step 3: Create a DataFrame for the amenities
amenities_df = pd.DataFrame(0, index=df.index, columns=unique_amenities)

# Step 4: Populate the DataFrame with 'Yes'/'No'
for amenity in unique_amenities:
    amenities_df[amenity] = df['amenities'].apply(lambda x: 'Yes' if amenity in x else 'No')

# Step 5: Concatenate the original DataFrame with the new amenities DataFrame
df = pd.concat([df, amenities_df], axis=1)
```
